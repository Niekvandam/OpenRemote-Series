Importing of Pandas and Pyplot, used for plotting the data

```python
import pandas as pd
import matplotlib.pyplot as plt

```

Load in the dataset from csv, plot the head of the dataset

```python
df = pd.read_csv("datapoints.csv", parse_dates=[0])
df.head()
```

Plot all the data, but only from BLOK61, create a mask of the desired values and plot them in pyplot, rounded up by date.

```python
mask = df.name.str.contains("BLOK61")
videolab = df[~mask]
df = videolab
plt.plot(df["value"])
plt.show()
df["timestamp"] = df["timestamp"].dt.round("D")
```

Figuring out how many days are in this dataset

```python
sorted_df = df["timestamp"].value_counts().sort_index()
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(sorted_df.index, sorted_df.values)
ax.tick_params(rotation=45)
```
